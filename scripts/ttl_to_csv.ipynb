ttl_to_csv.py

Convert a Turtle (.ttl) RDF file into a flat CSV with columns:
  subject, predicate, object

Usage: just set INPUT_TTL and OUTPUT_CSV below.

```python
# =========================
# CONFIG â€” EDIT HERE ONLY
# =========================
INPUT_TTL  = "output.schema+samples.ttl"
OUTPUT_CSV = "output_small.csv"
# =========================
```

```python
from rdflib import Graph
import csv

def main():
    g = Graph()
    g.parse(INPUT_TTL, format="turtle")
    print(f"Parsed {len(g)} triples from {INPUT_TTL}")

    with open(OUTPUT_CSV, "w", newline="", encoding="utf-8") as f:
        writer = csv.writer(f)
        writer.writerow(["subject", "predicate", "object"])
        for s, p, o in g:
            writer.writerow([str(s), str(p), str(o)])

    print(f"Wrote {len(g)} rows to {OUTPUT_CSV}")

if __name__ == "__main__":
    main()
```
